```python
class Kopek():

    def __init__(self,yas=5):
        self.yas = yas

    def kopekYasHesapla(self):
        print(self.yas * 7)
```

```python
benimKopek = Kopek()
```

```python
benimKopek.yas
```

```python
benimKopek.kopekYasHesapla()
```

```python
class Hayvan():
    def __init__(self):
        print("hayvan sınıfı çağrıldı")

    def method1(self):
        print("hayvan metodu 1 çağrıldı")

    def method2(self):
        print("hayvan sınıfı metod 2 çağrıldı")
```

```python
benimHAyvanim = Hayvan()
```

```python
benimHAyvanim.method1()
```

```python
benimHAyvanim.method2()
```

```python
class Kedi(Hayvan):
    def __init__(self):
        Hayvan.__init__(self)
        print("kedi sınıfı çağrıldı")

    def miyavla(self):
        print("miyav")
```

```python
keddy = Kedi()
```

```python
keddy.miyavla()
```

```python
keddy.method1()
```
