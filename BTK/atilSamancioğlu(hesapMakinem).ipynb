```python
def hesapla(a,b,islem):
    if islem not in "+-*/" :
        return "lütfen işlem seçiniz" 
    if islem == "+":
        return a+b
    if islem == "-":
        return a-b
    if islem == "*":
        return a*b
    if islem == "/":
        return a/b

    
    
```
