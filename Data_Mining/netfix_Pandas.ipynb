```python
import requests
import pandas as pd

api_key = "#####"  # kendi API anahtarını buraya yaz

movies = []

for page in range(1, 6):  # ilk 5 sayfa (100 film)
    url = (
        f"https://api.themoviedb.org/3/discover/movie"
        f"?api_key={api_key}"
        f"&with_watch_providers=8"
        f"&watch_region=TR"
        f"&sort_by=popularity.desc"
        f"&page={page}"
    )
    res = requests.get(url).json()
    for m in res['results']:
        movies.append({
            "Title": m['title'],
            "Rating": m['vote_average'],
            "Release Date": m['release_date'],
            "Overview": m['overview']
        })

df_movies = pd.DataFrame(movies)
df_movies.to_csv("netflix_movies.csv", index=False)
```

```python
tv_shows = []

for page in range(1, 6):  # ilk 5 sayfa (100 dizi)
    url = (
        f"https://api.themoviedb.org/3/discover/tv"
        f"?api_key={api_key}"
        f"&with_watch_providers=8"
        f"&watch_region=TR"
        f"&sort_by=popularity.desc"
        f"&page={page}"
    )
    res = requests.get(url).json()
    for s in res['results']:
        tv_shows.append({
            "Title": s['name'],
            "Rating": s['vote_average'],
            "First Air Date": s['first_air_date'],
            "Overview": s['overview']
        })

df_tv = pd.DataFrame(tv_shows)
df_tv.to_csv("netflix_tv_shows.csv", index=False)
```
