```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

import csv
```

```python
# Downloading imdb top 250 movie's data
url = 'https://www.imdb.com/chart/top/'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
response.text
```

```python
title = soup.find('strong', class_='poster-title').text.strip()
```

```python
imdb_rating = soup.find('span', class_='imdb').text.strip()
```

```python
title
```

```python
imdb_rating
```

```python
# Extract all titles and IMDb ratings
titles = [title.text.strip() for title in soup.find_all('strong', class_='poster-title')]
imdb_ratings = [rating.text.strip() for rating in soup.find_all('span', class_='imdb')]

```

```python
titles
```

```python
imdb_ratings
```

```python

for title in titles:
    print(title)
```

```python

data = []
for i in range(len(titles)):
    data.append((titles[i], imdb_ratings[i]))
```

```python
data
```

```python
csv_file = "Recommended-Movies.csv"
```

```python
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "IMDB"])
    writer.writerows(data)
    print(f"Data has been written to {csv_file} successfully.")

```
