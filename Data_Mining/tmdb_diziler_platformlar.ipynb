```python
import requests
import pandas as pd

api_key = "######"  # ← kendi key’ini buraya yaz

tv_shows = []

# 10 sayfa = 200 dizi
for page in range(1, 11):
    print(f"Fetching page {page}...")
    url = f"https://api.themoviedb.org/3/tv/top_rated?api_key={api_key}&language=en-US&page={page}"
    response = requests.get(url)
    data = response.json()

    for show in data['results']:
        tv_shows.append({
            "TV Show ID": show['id'],
            "Title": show['name'],
            "Rating": show['vote_average'],
            "genre_ids": show['genre_ids'],
            "Overview": show['overview'],
            "First Air Date": show['first_air_date']
        })

# DataFrame'e çevir
tv_df = pd.DataFrame(tv_shows)

# TV dizileri için tür isimlerini çek
genre_url = f"https://api.themoviedb.org/3/genre/tv/list?api_key={api_key}&language=en-US"
genre_data = requests.get(genre_url).json()
genre_map = {genre['id']: genre['name'] for genre in genre_data['genres']}

# genre_ids → tür isimleri
tv_df['Genres'] = tv_df['genre_ids'].apply(lambda ids: [genre_map.get(i, "Unknown") for i in ids])
tv_df.drop(columns=['genre_ids'], inplace=True)

# CSV olarak kaydet
tv_df.to_csv("tmdb_top_200_tv.csv", index=False)
print("Top 200 TV dizisi başarıyla kaydedildi!")

# İlk satırları göster
tv_df.head()
```
